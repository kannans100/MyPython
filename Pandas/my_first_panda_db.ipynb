```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
engine = create_engine('sqlite:///my_database.db')
```

```python
data = {
    'Name': ['Kannan', 'Karthik', 'Sukumar', 'David'],
    'Age': [25, 30, 35, 40],
    'City': ['Fremont', 'Paris', 'Delhi', 'Berlin']
}
df = pd.DataFrame(data)
print(df)
```

```python
df.to_sql('employees', con=engine, if_exists='replace', index=False)
```

```python
df = pd.read_sql('employees', con=engine)

# Or use a SQL query
query = "SELECT * FROM employees WHERE Age > 30"
df = pd.read_sql(query, con=engine)
```

```python
print(df)
```

```python
df['Age'] = df['Age'] * 1.1
```

```python
df.to_sql('employees_updated', con=engine, if_exists='replace', index=False)
```

```python
df = pd.read_sql('employees_updated', con=engine)

# Or use a SQL query
query = "SELECT * FROM employees_updated WHERE Age > 30"
df = pd.read_sql(query, con=engine)
print(df)
```

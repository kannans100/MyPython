```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Sales.csv")

# 2Ô∏è‚É£ Add a Total Sales column
df['Total Sales'] = df['Units Sold'] * df['Unit Price']

# 3Ô∏è‚É£ Group by product and region
sales_by_product = df.groupby('Product')['Total Sales'].sum()
sales_by_region = df.groupby('Region')['Total Sales'].sum()
```

```python
plt.figure(figsize=(8, 5))
sales_by_product.plot(kind='bar')
plt.title("Total Sales by Product")
plt.xlabel("Product")
plt.ylabel("Total Sales ($)")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sales_by_region.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title("Sales Distribution by Region")
plt.ylabel("")  # Hide y-label
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

sns.barplot(x='Product', y='Total Sales', data=df)
plt.title("Total Sales by Product (Seaborn Style)")
plt.show()
```

```python
plt.savefig("total_sales_by_product.png", dpi=300)
plt.show()
plt.figure(figsize=(6, 6))
sales_by_region.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
plt.title("Sales Distribution by Region")
plt.ylabel("")  # Hide y-axis label
plt.tight_layout()

# üëâ Save pie chart as JPG
plt.savefig("sales_by_region.jpg", dpi=300)
plt.show()
```

```python
from reportlab.lib.pagesizes import letter
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, Image
from reportlab.lib.styles import getSampleStyleSheet

pdf_file = "Sales_Report.pdf"
doc = SimpleDocTemplate(pdf_file, pagesize=letter)
styles = getSampleStyleSheet()
story = []

# --- PDF CONTENT ---
story.append(Paragraph("Sales Report", styles["Title"]))
story.append(Spacer(1, 20))

story.append(Paragraph("Dataset Summary:", styles["Heading2"]))
story.append(Paragraph(f"Total Rows: {len(df)}", styles["Normal"]))
story.append(Paragraph(f"Columns: {', '.join(df.columns)}", styles["Normal"]))
story.append(Spacer(1, 12))

# Product sales table
story.append(Paragraph("Total Sales by Product:", styles["Heading2"]))
for product, total in sales_by_product.items():
    story.append(Paragraph(f"‚Ä¢ {product}: ${total:,.2f}", styles["Normal"]))
story.append(Spacer(1, 12))

# Region sales table
story.append(Paragraph("Total Sales by Region:", styles["Heading2"]))
for region, total in sales_by_region.items():
    story.append(Paragraph(f"‚Ä¢ {region}: ${total:,.2f}", styles["Normal"]))
story.append(Spacer(1, 20))

# Add charts
story.append(Paragraph("Visualizations:", styles["Heading2"]))
story.append(Spacer(1, 10))
story.append(Image("total_sales_by_product.png", width=400, height=250))
story.append(Spacer(1, 20))
story.append(Image("sales_by_region.jpg", width=350, height=250))

# 6Ô∏è‚É£ Build the PDF
doc.build(story)

print(f"‚úÖ PDF report created successfully: {pdf_file}")
```

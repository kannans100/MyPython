```python
import pandas as pd

# Left DataFrame
df1 = pd.DataFrame({
    'id': [100, 200, 300,400],
    'name': ['Alice', 'Bob', 'Charlie','']
})

# Right DataFrame
df2 = pd.DataFrame({
    'id': [200, 300, 400],
    'score': [85, 90, 95]
})
```

```python
pd.merge(df1, df2, on='id', how='inner')
```

```python
pd.merge(df1, df2, on='id', how='left')
```

```python
pd.merge(df1, df2, on='id', how='right')
```

```python
pd.merge(df1, df2, on='id', how='outer')
```

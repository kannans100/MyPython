```python
import pandas as pd
data = {
    'Name': ['Alice', 'Bob', None, 'David', 'Eva', 'Alice'],
    'Age': ['25', '30', 'NaN', 'twenty', 22, '25'],
    'City': ['New York', 'Los Angeles', 'Chicago', 'New York', None, 'New York'],
    'Salary': [50000, 60000, 55000, None, 45000, 50000]
}

df = pd.DataFrame(data)
```

```python
print("Original DataFrame:\n", df)
```

```python
df.isnull().sum()
```

```python
df['Name'].fillna('Unknown', inplace=True)  # Fill missing names
df['City'].fillna('Unknown', inplace=True)  # Fill missing cities
df['Salary'].fillna(df['Salary'].mean(), inplace=True)  # Fill missing
```

```python
print(df)
```

```python
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
```

```python
print(df)
```

```python
df.drop_duplicates(inplace=True)
print(df)
```

```python
df.reset_index(drop=True, inplace=True)
df.dropna(subset=['Age'], inplace=True)
print(df)
```

```python
df['Name'] = df['Name'].str.strip()
df['City'] = df['City'].str.strip()
print(df)
```
